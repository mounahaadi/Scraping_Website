* Let's scrape the Subslikescript website
* Link of the website : `https://subslikescript.com/movies_letter-A?page=1`
* Let's write the very basic code to scrape the site

```python
# IMPORTING NECESSARY LIBRARIES
from bs4 import BeautifulSoup
import requests

website = "https://subslikescript.com/movie/A_1000000000000000_Ransom-4926482"

# GETTING RESPONSE AND CONTENT OF THE WEBSITE
response = requests.get(website)
soup = BeautifulSoup(response.text, 'lxml')

# CREATING A CONTAINER
container = soup.find('article', class_ ="main-article")
container
```

```python
# MOVIE NAME
movie_name = container.find('h1').text.replace(" - full transcript","")

# MOVIE SCRIPT
movie_script = container.find('div' , class_ = "full-script").get_text(separator='\n')
print(movie_script)
```

```python
with open(movie_name+'.txt', 'w') as f:
    f.write(movie_script)
    print("Created File - '" + movie_name + ".txt'" , "and added script inside it")
```
