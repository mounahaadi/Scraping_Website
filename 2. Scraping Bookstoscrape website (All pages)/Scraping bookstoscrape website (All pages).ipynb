```python
from bs4 import BeautifulSoup
import requests

website = "https://books.toscrape.com/index.html"
result = requests.get(website)

soup = BeautifulSoup(result.text, 'lxml')
soup
```

```python
container = soup.find('ol', class_ = "row")
container
```

```python
# BOOK NAME

headings_for_book_name = container.find_all('h3')
book_name = []

for i in headings_for_book_name:
    name = i.find('a')['title']
    print(type(i))
    book_name.append(name)
book_name
```

```python
# BOOK PRICE

headings_for_book_price = container.find_all('p', class_='price_color')
book_price = []

for i in headings_for_book_price:
    # Decode the text using UTF-8 encoding
    price = i.text.encode('iso-8859-1').decode('utf-8')
    book_price.append(price)
book_price
```

```python
# RATING

book_rating = []

rate = container.find_all('article', class_ = "product_pod")
for i in rate:
    rate = i.find('p')['class'][1]
    book_rating.append(rate)
book_rating
```

```python
for x, y, z in zip(book_name, book_price, book_rating):
    print(x,"|", y, "|", z)
```
